# filling the missing values in age by the mean values 

```python
train['Age'].mean()
print(train.isnull().sum())
```

```python
train['Age']=train['Age'].fillna(train['Age'].mean())
```

```python
train['Age'].mean()
train['Age'].mean()
print(train.isnull().sum())
```

# filling the mising values in Embarked values by mean values 

```python
train['Embarked']=train['Embarked'].fillna(train['Embarked'].mode()[0])
print(train.isnull().sum())
```

# convert Pclass to string(they are catrgorical)

```python
train.info()
```

```python
train['Pclass']=train['Pclass'].apply(str)
display(train.head(10))
```

# cperforming one hot encoding on  categorical features :

```python
print(train['Sex'].value_counts())
```

```python
print(train['Embarked'].value_counts())
```

```python
print(train.isnull().values.sum())
```

```python
# encoding the categorical values of 'Pclass' feature 
train_encoding = train.copy()
train_encoding = pd.get_dummies(train_encoding , columns =['Pclass'], prefix =['Pclass'])
# encoding the categorical values of  ' Embarked ' feature
train_coding =pd.get_dummies(train_encoding , columns = ['Embarked'],prefix = ['Embarked'])
# encoding the categorical calues of 'Sex 'feature Ã 
train_enc = pd.get_dummies (train_coding, columns=['Sex'])
```

```python
display(train_enc.head(10))
```

```python
labels = train_enc.pop('Survived')
display(labels.head(10))
```

```python
# we split our data 75% train and 25 % for test 
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(train_enc,labels,test_size=0.25)
```
